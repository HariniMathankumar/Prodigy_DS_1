```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn.csv')
df
```

```python
ages = df['Age'].dropna() 
```

```python
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in the Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders in the Population')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')  
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Chart')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='lightgreen')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
men_means = [20, 35, 30, 35]
women_means = [25, 32, 34, 20]
width = 0.35  
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(categories))
bar1 = ax.bar(x - width/2, men_means, width, label='Men', color='lightblue')
bar2 = ax.bar(x + width/2, women_means, width, label='Women', color='lightcoral')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Chart')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
bars1 = [20, 35, 30, 35]
bars2 = [25, 32, 34, 20]
plt.figure(figsize=(10, 6))
plt.bar(categories, bars1, color='lightblue', label='Bars1')
plt.bar(categories, bars2, bottom=bars1, color='lightcoral', label='Bars2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
start = [10, 15, 20, 25]
end = [30, 35, 40, 45]
plt.figure(figsize=(10, 6))
plt.barh(categories, width=end, left=start, color='lightgreen')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Floating Bar Chart')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
means = [20, 35, 30, 25]
errors = [2, 3, 4, 1]
plt.figure(figsize=(8, 6))
plt.bar(categories, means, yerr=errors, capsize=5, color='lightblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Error Bar Chart')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
np.random.seed(0)
ages = np.random.randint(18, 80, size=1000)  
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Ages')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
